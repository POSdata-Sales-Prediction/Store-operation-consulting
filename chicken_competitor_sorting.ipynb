```python
## 22.04.12 코드 전면 수정
import pandas as pd

# 크롤링한 치킨 데이터셋 Load
crawledData = pd.read_csv("../resources/chicken_gangnam.csv", sep='|', encoding="UTF8")

# 강남구 일반 음식점 인허가 데이터셋 Load
gangnam_food = pd.read_excel("../resources/rawdata/서울특별시강남구일반음식점인허가정보.xls")
```

```python
# 데이터 확인
crawledData
# 파일 0번행은 데이터 저작권 문구라 삭제함
gangnam_food = gangnam_food.iloc[1:]
```

```python
# 불필요한 컬럼 삭제
gangnam_food = gangnam_food[['사업장명','인허가일자', '인허가취소일자','영업상태명', '폐업일자', '도로명주소','업태구분명', '좌표정보(X)', '좌표정보(Y)']]
gangnam_food
```

```python
crawledData["상호명"]
```

```python
# 크롤링한 데이터에서 KEYWORD 추출
keyword = crawledData["상호명"]
keyword_new = []
for key in keyword:
    keyword_new.append(key.split(" ")[0])    
```

```python
# keyword 중복 제거 - 259개
keyword_new = list(set(keyword_new))
len(keyword_new)
```

```python
# keyword에 "치킨" 추가 - 총 260개
keyword_new.append("치킨")
keyword_new.sort()
print(keyword_new, len(keyword_new))
```

```python
# 1. 인허가 일자가 2019년 1월 31일 이전이며 2. 폐업일자가 2019년 3월 1일 이후이거나 폐업을 하지 않은 데이터 소팅
gangnam_food_2019 = gangnam_food[(pd.to_numeric(gangnam_food['인허가일자'])<20190131)]
gangnam_food_2019 = gangnam_food_2019[(pd.to_numeric(gangnam_food_2019['폐업일자'])>20190301) | gangnam_food_2019['폐업일자'].isna()] 
```

```python
gangnam_food_2019
```

```python
#keyword_new.remove("스페이스") # - 이상한 키워드 (ex.주식회사, 짱, 스페이스 삭제)
keyword_new_str = "|".join(keyword_new)
print(keyword_new_str)

gangnam_chicken_2019 = gangnam_food_2019[gangnam_food_2019["사업장명"].str.contains(keyword_new_str)]
```

```python
# 최종 sorting된 치킨집 - 309개
# 그 중 깐부치킨(30개) 삭제 - 최종 279개 
gangnam_chicken_competitor = gangnam_chicken_2019[gangnam_chicken_2019["사업장명"].str.contains("깐부치킨") ==False]
```

```python
gangnam_chicken_competitor.to_csv("../resources/extractedData/chicken_competitors_2019.csv", encoding= "utf8", index=False)
```
