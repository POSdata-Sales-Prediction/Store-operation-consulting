```python
## 22.04.12 코드 전면 수정
import pandas as pd

# 크롤링한 치킨 데이터셋 Load
crawledData = pd.read_csv("../resources/extractedData/gangnam_chicken_fin.csv", sep=',', encoding="UTF8")

# 강남구 일반 음식점 인허가 데이터셋 Load
gangnam_food = pd.read_excel("../resources/rawdata/서울특별시강남구일반음식점인허가정보.xls")
```

```python
# 데이터 확인
gangnam_food
# 파일 0번행은 데이터 저작권 문구라 삭제함
gangnam_food = gangnam_food.iloc[1:]
gangnam_food.columns
```

```python
# 불필요한 컬럼 삭제 - 관리번호: Primary key
gangnam_food = gangnam_food[['관리번호','사업장명','인허가일자', '인허가취소일자','영업상태명', '폐업일자', '도로명주소','업태구분명', '좌표정보(X)', '좌표정보(Y)']]
gangnam_food
```

```python
# 1. 인허가 일자가 2019년 1월 31일 이전이며 2. 폐업일자가 2019년 3월 1일 이후이거나 폐업을 하지 않은 데이터 소팅
gangnam_food_2019 = gangnam_food[(pd.to_numeric(gangnam_food['인허가일자'])<20190131)]
gangnam_food_2019 = gangnam_food_2019[(pd.to_numeric(gangnam_food_2019['폐업일자'])>20190301) | gangnam_food_2019['폐업일자'].isna()] 
```

```python
gangnam_food_2019
```

```python
# 현재 2022.4월 기준 강남구에 존재하는 치킨집 크롤링한 데이터 확인 
crawledData
```

```python
crawledData["상호명"]
```

```python
# 크롤링한 데이터에서 KEYWORD 추출
keyword = crawledData["상호명"]
keyword_new = []
for key in keyword:
    keyword_new.append(key.split(" ")[0])    
```

```python
# keyword 중복 제거 - 260개
keyword_new = list(set(keyword_new))
len(keyword_new)
```

```python
# keyword에 "치킨","비비큐","BHC","비에이치씨","b.h.c" 추가 - 총 265개
keyword_new.extend(["치킨","비비큐","BHC","비에이치씨","b.h.c"])
keyword_new = list(set(keyword_new))
keyword_new.sort()
print(keyword_new, len(keyword_new))
```

```python
#keyword_new.remove("스페이스") # - 이상한 키워드 (ex. 랜덤(Ran, 주식회사, 짱, 스페이스, 오늘은 삭제) --> 총 키워드 260개
keyword_new.remove("오늘은")
print(keyword_new, len(keyword_new))
```

```python
keyword_new_str = "|".join(keyword_new)
print(len(keyword_new))
print(keyword_new_str)

gangnam_chicken_2019 = gangnam_food_2019[gangnam_food_2019["사업장명"].str.contains(keyword_new_str)]

# 치킨집 키워드로 검색했을때, 332개 추출
print(gangnam_chicken_2019.shape)
```

```python
# 업태 구분명이 '호프/통닭', '통닭(치킨)' 인 데이터 Sort 
gangnam_hof_2019 = gangnam_food_2019[(gangnam_food_2019["업태구분명"] == "호프/통닭") | (gangnam_food_2019["업태구분명"] == "통닭(치킨)")]

# 앞서 키워드 안에 있는것에 맥주를 파는 집 추가로 더해줌 (총 269개)
hof_keyword = list(keyword_new)
hof_keyword.extend(["맥주","비어","호프","브롱스","브루잉","필스너하우스","통파이브","후라이드","닭"])
keyword_hof_str = "|".join(hof_keyword)
print(len(hof_keyword))
print(keyword_hof_str)
gangnam_hof_2019_set = gangnam_hof_2019[gangnam_hof_2019["사업장명"].str.contains(keyword_hof_str)]
```

```python
# 호프에서 223개 추가 추출
gangnam_hof_2019_set.shape
```

```python
# 호프데이터와 키워드 서칭 데이터 concat - outer join
gangnam_chicken_competitor = pd.concat([gangnam_chicken_2019, gangnam_hof_2019_set], join="outer")

# Primary key값으로 중복제거 - 총 555개
gangnam_chicken_competitor.drop_duplicates(['관리번호'])
gangnam_chicken_competitor
```

```python
# 최종 sorting된 치킨집 - 555개
# 그 중 깐부치킨 삭제 - 최종 508개 
gangnam_chicken_competitor = gangnam_chicken_competitor[gangnam_chicken_competitor["사업장명"].str.contains("깐부") ==False]
gangnam_chicken_competitor.shape
```

```python
# data 추출 전 마지막으로 사업장명 확인
print(list(gangnam_chicken_competitor["사업장명"]), len(gangnam_chicken_competitor))
```
