```python
import pandas as pd

# 행정동 매핑 데이터 load
code = pd.read_excel("../resources/rawdata/행정동코드_매핑정보_20200325.xlsx")

# 생활인구 데이터 load
active = pd.read_csv("../resources/rawdata/서울시 활동인구_201902.csv", index_col =False)
```

```python
# 데이터 확인 후 column 영문으로 수정 
code 
code.columns = code.iloc[0]
code = code[1:]
code
```

```python
# 강남구 코드만 sorting
code_gangnam = code[code["CT_NM"] == "강남구"]
code_gangnam
```

```python
# 생활인구 데이터 확인
active
```

```python
# 강남구 데이터만 소팅
active_gangnam = pd.DataFrame(columns=active.columns)
code = list(code_gangnam["H_DNG_CD"].astype(int))
for c in code :
    active_gangnam = active_gangnam.append(active[active["행정동코드"] == c])
```

```python
active_gangnam
```

```python
## 일평균 시간 구하기 
import numpy as np
active_gangnam = active_gangnam.iloc[:,[0,1,2,3]]
#active_gangnam_processed = pd.DataFrame(columns=["기준일ID","행정동코드","평균생활인구수","주중/주말"])
#active_gangnam_processed = active_gangnam.groupby(active_gangnam["기준일ID"],active_gangnam["행정동코드"])
active_gangnam_processed = pd.DataFrame(active_gangnam.groupby(["기준일ID","행정동코드"], as_index=False)["총생활인구수"].mean())
```

```python
# 컬럼명 변경
active_gangnam_processed.columns= ["기준일ID","행정동코드","일별평균생활인구수"]

# 요일 컬럼 추가
import datetime
active_gangnam_processed["기준일ID"] = pd.to_datetime(active_gangnam_processed["기준일ID"], format="%Y%m%d")
# 0-월, 1-화 ... 5-토,6-일
active_gangnam_processed["요일"] = active_gangnam_processed["기준일ID"].dt.weekday
```

```python
# 일별 평균 생활 인구수 기초 통계량
ps = active_gangnam_processed["일별평균생활인구수"]
print("개수:",ps.count())
print("합계:",ps.sum())
print("평균:",ps.mean())
print("분산:",ps.var())
print("표준편차:",ps.std())
print("중간값:",ps.median())
print("최솟값:",ps.min())
print("최댓값:",ps.max())
print("10%:",ps.quantile(0.1))
print("50%:",ps.quantile(0.5))
print("70%:",ps.quantile(0.7))
```

```python
# 일평균 생활 인구수 plot
import matplotlib.pyplot as plt

# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

for c in code:
    y = active_gangnam_processed[active_gangnam_processed["행정동코드"] == c]["일별평균생활인구수"]
    x = active_gangnam_processed[active_gangnam_processed["행정동코드"] == c]["기준일ID"]
    name = code_gangnam[code_gangnam["H_DNG_CD"] == str(c)]["H_DNG_NM"].values[0]+" 일별 평균 생활인구수"
    print(name)
    ps = active_gangnam_processed["일별평균생활인구수"]
    # 기초 통계량 출력
    print("개수:",y.count())
    print("합계:",y.sum())
    print("평균:",y.mean())
    print("분산:",y.var())
    print("표준편차:",y.std())
    print("중간값:",y.median())
    print("최솟값:",y.min())
    print("최댓값:",y.max())
    print("10%:",y.quantile(0.1))
    print("50%:",y.quantile(0.5))
    print("70%:",y.quantile(0.7))
    
    # 플롯 생성
    plt.plot(x,y)
    plt.title(name, fontsize=20)
    plt.gcf().subplots_adjust(bottom=0.30)
    plt.xticks(rotation= 90)
    
    # 플롯 저장
    plt.savefig("../resources/plot/"+name+".png")
    plt.show()
    
```

```python
# 주거 인구 데이터 Load
live_people = pd.read_excel("../resources/rawdata/201902_인구 및 세대현황(전체).xlsx")
# 불필요한 컬럼 삭제
live_people = live_people.iloc[:,:6]

# 데이터 확인 - 행정동 데이터 에러 해결
live_people["행정동"] = live_people["행정동"].str.replace(" ","")
live_people
```

```python
# 주중 평균과 주말 평균 구하기
active_gangnam_week = pd.DataFrame(columns=["행정동","인구수","주중평균활동인구","주말평균활동인구","일평균활동인구","유동인구비율"])

# 2월 5일(설 당일) 제거
idx = active_gangnam_processed[active_gangnam_processed["기준일ID"] == "2019-02-05"].index
active_gangnam_processed = active_gangnam_processed.drop(idx)

# 행정동별 주중, 주말, 일 평균 생활 인구수와 인구수, 비율 추출
for c in code:
    name = code_gangnam[code_gangnam["H_DNG_CD"] == str(c)]["H_DNG_NM"].values[0]
    lp = live_people[live_people["행정동"] == name]["계"].values[0]
    day = round(active_gangnam_processed[(active_gangnam_processed["행정동코드"] == c) & (5> active_gangnam_processed["요일"])]["일별평균생활인구수"].mean(),2)
    dayoff = round(active_gangnam_processed[(active_gangnam_processed["행정동코드"] == c) & (active_gangnam_processed["요일"] > 4)]["일별평균생활인구수"].mean(),2)
    daily = round(active_gangnam_processed[(active_gangnam_processed["행정동코드"] == c)]["일별평균생활인구수"].mean(),2)
    ratio = round(daily/lp, 2)
    active_gangnam_week = active_gangnam_week.append({"행정동":name, "인구수":lp, "주중평균활동인구":day, "주말평균활동인구":dayoff,
                                                      "일평균활동인구": daily, "유동인구비율":ratio},ignore_index=True)

active_gangnam_week.to_excel("../resources/extractedData/강남구 행정동별 인구 feature.xlsx", index=False) 
```
