```python
import pandas as pd

# 예린, 서연 파일 경로 (22.06.05 서연 파일 경로 예린과 통일)
#sy_file = "../../rawdata/"
sy_file = "../../../resources/rawdata/"

# 변수명이 애매하므로 수정 (22.06.15 예린)
store_info = pd.read_csv(sy_file+'매장데이터_수정.csv', encoding='UTF8')
sales = pd.read_csv(sy_file+'거래데이터_full.csv', encoding='UTF8')
```

```python
store_info 
```

```python
from collections import Counter

#본사코드 별 등장 횟수 카운트
CD = store_info['본사코드']
head_counter = Counter(CD)

head_counter
```

```python
#본사코드 중 등장 횟수가 5 이상인 것만 head_counter_sorted에 저장
head_counter_sorted = dict()
for key in head_counter.keys() :     
        if head_counter[key] >= 5 : head_counter_sorted[key] = head_counter[key]

head_counter_sorted
```

```python
# store_info 에서 본사코드가 0000이면 개인 사업자임
# 따라서, 0000 제거하고 프랜차이즈만 소팅
idx = store_info[store_info['본사코드'] == '0000'].index
franchise_store_info = store_info.drop(idx)
franchise_store_info
```

```python
# 매머드커피 매장코드 리스트 추출
mamm_CD = list(store_info[store_info['본사코드']=="HDSF"]['매장코드'])

# 매머드커피 데이터만 저장할 dataframe 생성
mamm_df = pd.DataFrame(sales[sales['매장코드']==mamm_CD[0]])  #매장코드 중 첫 번째인 것의 데이터만 추출해 데이터프레임 생성

#매장코드 중 첫 번째를 제외한 나머지를 이전에 생성한 데이터프레임에 추가
for cd in mamm_CD[1:] : 
    mamm_df = mamm_df.append(sales[sales['매장코드']==cd])

mamm_df
```

```python
# 매머드커피의 소분류별 메뉴명 출력
for i in mamm_df['소분류명'].unique():
    print(i)
    print(mamm_df[mamm_df['소분류명'] == i]['상품명'].unique())
```
