```python
import pandas as pd


#엑셀 파일을 csv 파일로 변환
ex = pd.read_excel('../rawdata/매장데이터_수정.xls')
ex.to_csv("../rawdata/매장데이터_수정.csv", index = False)

data = pd.read_csv('../rawdata/매장데이터_수정.csv', encoding='UTF8')
```

```python
from collections import Counter

#본사코드 별 등장 횟수 카운트
column_CD = data['본사코드']
counter = Counter(column_CD)

counter
```

```python
del counter['0000']     #'0000'의 경우, 프랜차이즈가 아님을 뜻하므로 삭제

#본사코드 중 등장 횟수가 5 이상인 것만 cd5에 저장
cd5 = dict()
for key in counter.keys() :     
        if counter[key] >= 5 : cd5[key] = counter[key]

cd5
```

```python
len(data[data['본사코드']=="HBAS"]) #기존데이터에서 본사코드가 HBAS(깐부치킨)의 등장 횟수 체크
```

```python
data_full = pd.read_csv('../rawdata/거래데이터_full.csv', encoding='UTF8')
```

```python
data_full
```

```python
CD = list(data[data['본사코드']=="HBAS"]['매장코드'])   #깐부치킨 매장의 매장코드 추출
```

```python
df = pd.DataFrame(data_full[data_full['매장코드']==CD[0]])  #매장코드 중 첫 번째인 것의 데이터만 추출해 데이터프레임 생성

#매장코드 중 첫 번째를 제외한 나머지를 이전에 생성한 데이터프레임에 추가
for cd in CD[1:] : df = df.append(data_full[data_full['매장코드']==cd])

df
```

```python
df.to_csv('../processedData/깐부치킨_거래데이터.csv', index=False)    #csv로 저장
```
