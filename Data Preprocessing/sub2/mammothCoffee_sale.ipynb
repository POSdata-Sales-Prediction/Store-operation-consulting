### 매머드커피 매장별 일자, 시간대별로 총매출 추출 코드

```python
import pandas as pd
mamm = pd.read_csv("../../../resources/extractedData/매머드_거래데이터.csv", encoding="utf8")
```

```python
# column : 매장코드, 매장명, 영업일자, (결제일시->)시간대, 매출
df = mamm[["매장코드", "매장명", "영업일자", "결제일시", "총판매금액"]]
```

```python
# 결제일시 datetime으로 포매팅
df["결제일시"] = df["결제일시"].astype("str")
df["결제일시"] = pd.to_datetime(df["결제일시"])

# 시간대 추출
df["시간대"] = df["결제일시"].dt.hour
```

```python
df
```

```python
# 매장별 시간대별 매출
df_sum = pd.DataFrame({"매출" : df.groupby(["매장코드", "매장명", "영업일자", "시간대"])['총판매금액'].sum()}).reset_index()
df_sum.to_csv("../../../resources/processedData/매머드_매장별 시간대별 매출.csv", index=False , encoding="utf8")
df_sum
```
