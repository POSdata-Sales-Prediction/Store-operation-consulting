```python
## 22.05.04 
# 반경 500m 로 feature 재 생산이라 서초구, 송파구 경쟁업체 데이터 필요해짐

import pandas as pd

# 이전 강남구 crawling 코드에서 추출한 keywords 데이터 load
file = open("../resources/rawdata/chicken_keywords.txt", 'r', encoding="utf8")
keywords = file.read()

# 송파구, 서초구 일반 음식점 인허가 데이터셋 load
songpa_food = pd.read_excel("../resources/rawdata/서울시 송파구 일반음식점 인허가 정보.xls", index_col=False)
seocho_food = pd.read_excel("../resources/rawdata/서울시 서초구 일반음식점 인허가 정보.xls", index_col=False)
```

```python
# keyword를 list로 변환
keywords = keywords.replace("'","")

keyword = keywords.split(", ")
keyword
```

```python
songpa_food
```

```python
# 불필요한 컬럼 삭제 - 관리번호: Primary key
songpa_food = songpa_food[['관리번호','사업장명','인허가일자', '인허가취소일자','영업상태명', '폐업일자', '도로명주소','업태구분명', '좌표정보(X)', '좌표정보(Y)']]
seocho_food = seocho_food[['관리번호','사업장명','인허가일자', '인허가취소일자','영업상태명', '폐업일자', '도로명주소','업태구분명', '좌표정보(X)', '좌표정보(Y)']]
```

```python
# 1. 인허가 일자가 2019년 1월 31일 이전이며 2. 폐업일자가 2019년 3월 1일 이후이거나 폐업을 하지 않은 데이터 소팅
songpa_food_2019 = songpa_food[(pd.to_numeric(songpa_food['인허가일자'])<20190131)]
songpa_food_2019 = songpa_food_2019[(pd.to_numeric(songpa_food_2019['폐업일자'])>20190301) | songpa_food_2019['폐업일자'].isna()] 

seocho_food_2019 = seocho_food[(pd.to_numeric(seocho_food['인허가일자'])<20190131)]
seocho_food_2019 = seocho_food_2019[(pd.to_numeric(seocho_food_2019['폐업일자'])>20190301) | seocho_food_2019['폐업일자'].isna()] 
```

```python
# 데이터 확인
seocho_food_2019["사업장명"].isna().sum()
```

```python
# chicken keyword에 검색되는 데이터 추출
keyword_new_str = "|".join(keyword)
print(len(keyword))
print(keyword_new_str)

songpa_chicken_2019 = songpa_food_2019[songpa_food_2019["사업장명"].str.contains(keyword_new_str, na=False)]
seocho_chicken_2019 = seocho_food_2019[seocho_food_2019["사업장명"].str.contains(keyword_new_str, na=False)]

# 치킨집 키워드로 검색했을때, 332개 추출
print(songpa_chicken_2019.shape)
print(seocho_chicken_2019.shape)
```

```python
# 업태 구분명이 '호프/통닭', '통닭(치킨)' 인 데이터 Sort 
songpa_hof_2019 = songpa_food_2019[(songpa_food_2019["업태구분명"] == "호프/통닭") | (songpa_food_2019["업태구분명"] == "통닭(치킨)")]
seocho_hof_2019 = seocho_food_2019[(seocho_food_2019["업태구분명"] == "호프/통닭") | (seocho_food_2019["업태구분명"] == "통닭(치킨)")]

# 앞서 키워드 안에 있는것에 맥주를 파는 집 추가로 더해줌 (총 269개)
hof_keyword = list(keyword)
hof_keyword.extend(["맥주","비어","호프","브롱스","브루잉","필스너하우스","통파이브","후라이드","닭"])
keyword_hof_str = "|".join(hof_keyword)
print(len(hof_keyword))
print(keyword_hof_str)

songpa_hof_2019_Set = songpa_hof_2019[songpa_hof_2019["사업장명"].str.contains(keyword_hof_str)]
seocho_hof_2019_Set = seocho_hof_2019[seocho_hof_2019["사업장명"].str.contains(keyword_hof_str)]


# 호프에서 223개 추가 추출
print("songpa: ",songpa_hof_2019_Set.shape)
print("seocho: ",seocho_hof_2019_Set.shape)
```

```python
# 호프데이터와 키워드 서칭 데이터 concat - outer join
songpa_chicken_competitor = pd.concat([songpa_chicken_2019, songpa_hof_2019_Set], join="outer")
seocho_chicken_competitor = pd.concat([seocho_chicken_2019, seocho_hof_2019_Set], join="outer")

# Primary key값으로 중복제거 - 총 555개
songpa_chicken_competitor.drop_duplicates(['관리번호'])
seocho_chicken_competitor.drop_duplicates(['관리번호'])
```

```python
# 최종 sorting된 치킨집 - 555개
# 그 중 깐부치킨 삭제 - 최종 508개 
songpa_chicken_competitor = songpa_chicken_competitor[songpa_chicken_competitor["사업장명"].str.contains("깐부") ==False]
seocho_chicken_competitor = seocho_chicken_competitor[seocho_chicken_competitor["사업장명"].str.contains("깐부") ==False]

print("songpa: ",songpa_chicken_competitor.shape)
print("seocho: ",seocho_chicken_competitor.shape)
```
