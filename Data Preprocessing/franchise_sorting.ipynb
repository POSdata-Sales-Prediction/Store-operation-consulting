```python
# Import Moudules
import pandas as pd 
import matplotlib.pyplot as plt

shop_info19 = pd.read_excel('../../resources/rawdata/매장데이터_수정.xls')
```

```python
shop_info19_dropped = shop_info19[shop_info19['구분']!='비식당']
```

```python
franchise = shop_info19_dropped[shop_info19_dropped['본사코드'] != "0000"]
franchise
```

```python
from collections import Counter
data = Counter(franchise['본사코드'])
data
```

```python
# 매장이 5개 이상인 점포만 sorting 후 이름과 종목명 출력
pgm_lang_val_reverse = sorted(data.items(), 

                              reverse=True, 

                              key=lambda item: item[1])

franchise_sorted = dict()

for key, value in pgm_lang_val_reverse:
    if (value > 4) : 
        name = franchise[franchise['본사코드'] == key]['매장명'].tolist()
        category = franchise[franchise['본사코드'] == key]['종목명'].tolist()
        franchise_sorted[key] =  franchise[franchise['본사코드'] == key]['매장코드'].tolist()
        print(key, ":", value, name[0], category[0])
        
```
