### 깐부치킨 메뉴 전처리

```python
## import modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

## load data
gganbu_raw = pd.read_csv("../../../resources/processedData/깐부치킨_거래데이터.csv", encoding="utf8")
```

```python
gganbu_code = gganbu_raw[["매장코드","매장명"]].drop_duplicates()
gganbu_code.to_csv("../../../resources/processedData/깐부치킨_매장코드.csv", encoding="utf8", index=False)
```

```python
## 각 메뉴명 1차 클렌징
# 고추간장치킨 -> 고추간장/ 마늘간장(순살), 마늘간장치킨 -> 마늘간장
# 순살떡볶이-> 순살 국물 떡볶이/ 순살스윗치킨 -> 순살스윗
# 후라이드양념반반 -> 후라이드 반반
# 깐부불사조 -> 불사조
# 코울슬러, 코오슬로추가 -> 코울슬로
# 카스, 카스(병)..-> 카스(병)
# 클라우드, 클라우드 맥주 -> 클라우드(병)
# 코르넨버그 1664 블랑 -> 1644블랑
# 공백, 쉼표 제거

gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("고추간장치킨","고추간장")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("마늘간장(순살)","마늘간장")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("마늘간장치킨","마늘간장")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("순살떡볶이","순살국물떡볶이")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("순살스윗치킨","순살스윗")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("후라이드양념반반","후라이드 반반")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("깐부불사조","불사조")

gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("코울슬러","코울슬로")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("코오슬로추가","코울슬로")

gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("카스","카스(병)")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("카스(병)..","카스(병)")

gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("클라우드","클라우드(병)")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("클라우드 맥주","클라우드(병)")
gganbu_raw["상품명"] = gganbu_raw["상품명"].replace("생맥주1000cc(포장)..","생맥주1000cc(포장)")

gganbu_raw["상품명"] = gganbu_raw["상품명"].str.replace(" ","")
gganbu_raw["상품명"] = gganbu_raw["상품명"].str.replace(",","")
```

```python
gganbu_raw.to_csv("../../../resources/processedData/깐부치킨_청담1x_거래데이터_1차_cleansing.csv",index=False, encoding="utf8")
```

```python
## 판매 매장 갯수 
store_count = pd.DataFrame({'판매매장' : gganbu_raw.groupby(["소분류명","상품명"])["매장코드"].nunique()}).reset_index()
store_count.to_csv("../../../resources/processedData/깐부치킨_판매 매장 갯수.csv",index=False, encoding="utf8")
store_count
```

```python
## 매장별 메뉴 판매 갯수 
menu_store_count = pd.DataFrame({'판매개수' : gganbu_raw.groupby(["상품명","판매단가","매장코드","매장명"] )["판매수량"].sum()}).reset_index()
menu_store_count.to_csv("../../../resources/processedData/깐부치킨_매장별_메뉴_판매 개수.csv",index=False, encoding="utf8")
menu_store_count
```

```python
## 메뉴별 총 판매 갯수
menu_whole_count = pd.DataFrame({'총판매개수' : menu_store_count.groupby(["상품명"] )["판매개수"].sum()}).reset_index()
menu_whole_count.to_csv("../../../resources/processedData/깐부치킨_메뉴별_총 판매 개수.csv",index=False, encoding="utf8")
menu_whole_count
```
