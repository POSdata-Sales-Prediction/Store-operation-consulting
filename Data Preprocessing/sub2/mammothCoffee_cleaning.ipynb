```python
import pandas as pd

mamm = pd.read_csv("../../../resources/extractedData/매머드_거래데이터_1차.csv", encoding="utf8")
```

```python
mamm
```

```python
## 테이블 번호, 총판매금액, 상품코드, 판매단가 삭제

del mamm['테이블번호']
del mamm['총판매금액']
del mamm['상품코드']
del mamm['판매단가']

mamm
```

```python
## 상품명 사이즈 표기 삭제 및 상품명 통일
# S화이트초콜릿라떼(hot) -> 화이트초코라떼(hot)
# S화이트초콜릿라떼(ice) -> 화이트초코라떼(ice)
# S화이트초콜릿프라페 -> 화이트초콜릿프라페

mamm['상품명'] = mamm['상품명'].replace('S화이트초콜릿라떼(hot)', '화이트초코라떼(hot)')
mamm['상품명'] = mamm['상품명'].replace('S화이트초콜릿라떼(ice)', '화이트초코라떼(ice)')
mamm['상품명'] = mamm['상품명'].replace('S화이트초콜릿프라페', '화이트초콜릿프라페')
```

```python
mamm.to_csv('../../../resources/extractedData/매머드_거래데이터_2차.csv', index=False)
```

```python
mamm_sorted = mamm[(mamm["소분류명"]!="Add") & (mamm["소분류명"]!="Dessert")]
```

```python
## 판매 매장 갯수 
# mamm_sorted-> 서연이 df로 변경
# 서연이 파일 경로 수정 -> 파일명: 매머드커피_2차_판매 매장 갯수
store_count = pd.DataFrame({'판매매장' : mamm_sorted.groupby(["소분류명","상품명"])["매장코드"].nunique()}).reset_index()
store_count.to_csv("../../../resources/processedData/매머드커피_판매 매장 갯수_2차.csv",index=False, encoding="utf8")
store_count
```

```python
## 2. 매장별 메뉴 판매 갯수 
# mamm_sorted-> 서연이 df로 변경
# 서연이 파일 경로 수정 -> 파일명: 매머드커피_2차_판매 매장 갯수
menu_store_count = pd.DataFrame({'판매개수' : mamm_sorted.groupby(["상품명","매장코드","매장명"] )["판매수량"].sum()}).reset_index()
menu_store_count.to_csv("../../../resources/processedData/매머드커피_매장별_메뉴_판매 개수_2차.csv",index=False, encoding="utf8")
menu_store_count
```

```python
## 3. 메뉴별 총 판매 갯수
# menu_store_count-> 이거 그냥 두면 됨 위에서 만든 데이터 프레임에서 재 추출하는 코드라 가만히 두기!
menu_whole_count = pd.DataFrame({'총판매개수' : menu_store_count.groupby(["상품명"] )["판매개수"].sum()}).reset_index()
menu_whole_count.to_csv("../../../resources/processedData/매머드커피_메뉴별_총 판매 개수_2차.csv",index=False, encoding="utf8")
menu_whole_count
```
