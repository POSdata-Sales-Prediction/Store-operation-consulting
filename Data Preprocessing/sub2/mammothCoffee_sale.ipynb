### 매머드커피 매장별 일자, 시간대별로 총매출 추출 코드

```python
import pandas as pd
# mamm = pd.read_csv("../../../resources/extractedData/매머드_거래데이터.csv", encoding="utf8")
mamm = pd.read_csv("../../../resources/extractedData/매머드_거래데이터_주말 추가.csv", encoding="utf8")
```

```python
mamm
```

```python
# column : 매장코드, 매장명, 영업일자, (결제일시->)시간대, 매출
mamm_date = mamm[["매장코드", "매장명", "영업일자", "결제일시", "주중/주말", "총판매금액"]]
```

```python
# 결제일시 datetime으로 포매팅
mamm_date["결제일시"] = mamm_date["결제일시"].astype("str")
mamm_date["결제일시"] = pd.to_datetime(mamm_date["결제일시"])

# 시간대 추출
mamm_date["시간대"] = mamm_date["결제일시"].dt.hour
```

```python
mamm_date
```

```python
# 매장별 시간대별 매출
mamm_sum = pd.DataFrame({"매출" : mamm_date.groupby(["매장코드", "매장명", "영업일자", "시간대", "주중/주말"])['총판매금액'].sum()}).reset_index()
mamm_sum.to_csv("../../../resources/processedData/매머드_매장별 시간대별 매출_주말 추가.csv", index=False , encoding="utf8")
mamm_sum
```

시간대별 평균 매출 구하기 
- 빈 시간대 값 채우기
- 주중/주말 별 시간대 치환

```python
import pandas as pd

mamm_sum = pd.read_csv("../../../resources/processedData/매머드_매장별 시간대별 매출_주말 추가.csv", encoding="utf8")

# 주중/주말 문자열로 변수형 변경
# mamm_sum["주중/주말"] = mamm_sum["주중/주말"].astype("str")
```

```python
mamm_sum
```

```python
# 매장별 주중/주말 영업 시간(min, max) / 영업일자 및 요일 확인

date = ["금", "토", "일", "월", "화", "수", "목"]
for i in mamm_sum["매장명"].unique():
    data = mamm_sum[mamm_sum["매장명"] == i] 

    # 매장명, 주중/주말 영업 시간(min, max)
    print("%s 주중: %s %s 주말: %s %s " % (i, data[data["주중/주말"] == 0]["시간대"].min(), data[data["주중/주말"] == 0]["시간대"].max(), data[data["주중/주말"] == 1]["시간대"].min(), data[data["주중/주말"] == 1]["시간대"].max()))
    # 영업일자 및 요일 확인
    for j in data["영업일자"].unique():
        print(j, date[int(j[8:])%7-1])
    print()
```

```python
## 시간대 중 없는 시간대의 매출 0으로 row 추가 (주중, 주말 나누기) 

for code in mamm_sum["매장코드"].unique(): 
    store = mamm_sum[mamm_sum["매장코드"]==code]
    
    for date in store["영업일자"].unique():
        store_date = store[store["영업일자"] == date]

        # 영업 시간 중 데이터가 없는 시간대를 매출 0으로 row 추가
        store_date_hour = list(store_date["시간대"])    # 실제 데이터가 있는 시간대
        for h in range(store_date["시간대"].min(), store_date["시간대"].max()+1) : 
            if h not in store_date_hour:    # 영업시간 중, 데이터가 없는 시간대가 있다면 행 추가
                mamm_sum = mamm_sum.append({"매장코드" : code, "매장명" : store_date["매장명"].values[0], "영업일자" : date, "시간대" : h, "주중/주말" : store_date["주중/주말"].values[0], "매출" : 0}, ignore_index=True)

# 매장코드, 영업일자, 시간대 기준 오름차순 정렬
mamm_sum = mamm_sum.sort_values(["매장코드", "영업일자", "시간대"], ignore_index=True)
mamm_sum
```

```python
## 주중/주말별 시간대 1~4로 치환

# 주중, 주말로 데이터프레임 나누기
mamm_weekday = mamm_sum[mamm_sum["주중/주말"] == 0]
mamm_weekend = mamm_sum[mamm_sum["주중/주말"] == 1]

# 주중 시간대 치환 / 1 : (6, 7, 8, 9, 10), 2: (11, 12, 13, 14), 3: (15, 16, 17, 18), 4: (19, 20, 21, 22, 23)
mamm_weekday = mamm_weekday.replace([6, 7, 8, 9, 10], 1)
mamm_weekday = mamm_weekday.replace([11, 12, 13, 14], 2)
mamm_weekday = mamm_weekday.replace([15, 16, 17, 18], 3)
mamm_weekday = mamm_weekday.replace([19, 20, 21, 22, 23], 4)

# 주말 시간대 치환 / 1: (7, 8, 9, 10, 11), 2: (12, 13, 14, 15), 3: (16, 17, 18, 19), 4: (20, 21, 22, 23)
mamm_weekend = mamm_weekend.replace([7, 8, 9, 10, 11], 1)
mamm_weekend = mamm_weekend.replace([12, 13, 14, 15], 2)
mamm_weekend = mamm_weekend.replace([16, 17, 18, 19], 3)
mamm_weekend = mamm_weekend.replace([20, 21, 22, 23], 4)

# 하나의 데이터프레임으로 합치기 및 정렬
mamm_total = pd.concat([mamm_weekday,mamm_weekend], ignore_index=True).sort_values(["매장코드", "영업일자", "시간대"])
mamm_total
```

```python
## 시간대별 평균 매출 구하기 및 데이터프레임 추출
# columns : 매장코드_영업일자_시간대(key) / 매장코드 / 매장명 / 영업일자 / 시간대 / 매출

# 시간대별 평균 매출 추출
mamm_final2 = pd.DataFrame({"평균매출" : round(mamm_total.groupby(["매장코드", "매장명", "영업일자", "주중/주말", "시간대"])["매출"].mean(), 2)}).reset_index()
# key(매장코드_영업일자_시간대) 생성
mamm_final2["매장코드_영업일자_시간대"] = mamm_final2.apply(lambda x: "%s_%s_%s" %(x["매장코드"], x["영업일자"], x["시간대"]), axis='columns')

mamm_final2 = mamm_final2[["매장코드_영업일자_시간대", "매장코드", "매장명", "영업일자", "주중/주말", "시간대", "평균매출"]]
mamm_final2.to_csv("../../../resources/processedData/매머드_매장별 시간대별 평균매출.csv", index=False , encoding="utf8")
mamm_final2
```
