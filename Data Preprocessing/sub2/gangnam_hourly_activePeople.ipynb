### 강남구, 송파구, 서초구 시간대별 활동 인구 추출 코드
### feature중 시간대별 활동인구를 추출하기 위함

```python
## Import modules & Load Dataset
import pandas as pd

# 행정동 매핑 데이터 load
code = pd.read_excel("../../../resources/rawdata/행정동코드_매핑정보_20200325.xlsx")

# 생활인구 데이터 load
active = pd.read_csv("../../../resources/rawdata/서울시 활동인구_201902.csv", index_col =False)

# 입지특성 데이터 load
geo_feature = pd.read_csv("../../../resources/extractedData/매머드_입지특성_최종.csv", index_col =False)
geo_feature = geo_feature.iloc[:,:20]
```

```python
# 행정동 매핑 데이터 확인 후 column 영문으로 수정 
code.columns = code.iloc[0]
code = code[1:]
code['H_SDNG_CD'] = code['H_SDNG_CD'].astype(str)
code
```

```python
# 입지특성 feature에 나온 행정동만 sorting
dong_code = list(geo_feature.columns[3:])
keyword_new_str = "|".join(dong_code)

code_sorted = code[code["H_DNG_CD"].str.contains(keyword_new_str)]
code_sorted
```

```python
## 22.06.28 입지특성 column 수정 - 잘못된 코드로 되어있었음................. 
# 상업지역 면적 반올림
geo_feature["상업지역 면적"] = round(geo_feature["상업지역 면적"], 2)

# 컬럼명 수정
#geo_feature.rename(columns=, inplace=True)
```

```python
# 생활인구 데이터 확인
active
```

```python
# 생활인구 데이터 중 강남구 데이터만 소팅
active_dong = pd.DataFrame(columns=active.columns)
code = list(code_sorted["H_DNG_CD"].astype(int))
for c in code :
    active_dong = active_dong.append(active[active["행정동코드"] == c])
active_dong
```

```python
# 생활 인구 데이터 중 총 생활 인구수만 사용
active_dong = active_dong.iloc[:,[0,1,2,3]]

# 06시부터 23시까지 데이터만 사용 
active_dong = active_dong.loc[active_dong[active_dong["시간대구분"]>5].index,:]
active_dong
```

```python
# 시간대별 생활인구 수 리스트 만들기
pivot_active = pd.pivot_table(active_dong, values="총생활인구수", index=["기준일ID","시간대구분"], columns="행정동코드")
hourly_active = pivot_active.reset_index()

# 입지특성 df와 컬럼 순서 동일하게 변경
hourly_active = hourly_active.reindex(columns=["기준일ID","시간대구분"]+list(map(int, dong_code)))
hourly_active
```

```python
import numpy as np
# Feature 뽑기
geo_people_feature = pd.DataFrame(columns=[ "매장코드", "매장명", "상업지역 면적","기준일ID","시간대구분", "총생활인구수"])
for store_idx in range(len(geo_feature)):
    # 매장 정보 뽑기 - 매장코드, 매장명, 곱할 숫자
    store_info = geo_feature.iloc[store_idx, :3].values
    store_ratio = geo_feature.iloc[store_idx, 3:].values
    
    # 시간대 별 활동인구 다 더하기
    for hourly_idx in range(len(hourly_active)):
        date_hour = hourly_active.iloc[hourly_idx, :2].values
        whole_people = sum(store_ratio * hourly_active.iloc[hourly_idx, 2:].values)
        geo_people_feature = geo_people_feature.append(dict(zip(geo_people_feature.columns, np.hstack([store_info,date_hour,whole_people]))) ,ignore_index=True)
```

```python
geo_people_feature
```

```python
## 주중, 주말 데이터 매핑하기
# 주중 - 0, 주말 (+[설날] 19-02-04, 19-02-05, 19-02-06) - 1

# 요일 컬럼 추가
import datetime
geo_people_feature["기준일ID"] = pd.to_datetime(geo_people_feature["기준일ID"], format="%Y%m%d")

# 0-월, 1-화 ... 5-토,6-일
geo_people_feature["요일"] = geo_people_feature["기준일ID"].dt.weekday

# 요일 > 4 (주말) - 1, 요일 < 5 (주중) - 0
geo_people_feature["주중/주말"] = 0
geo_people_feature.loc[geo_people_feature[geo_people_feature["요일"] > 4].index,["주중/주말"]] = 1

# 설날 19-02-04, 19-02-05, 19-02-06 주말 취급
query = (geo_people_feature["기준일ID"] == "2019-02-04") | (geo_people_feature["기준일ID"] == "2019-02-05") | (geo_people_feature["기준일ID"] == "2019-02-06")
idx = geo_people_feature[query].index
geo_people_feature.loc[idx,["주중/주말"]] = 1

# 요일 컬럼 삭제
geo_people_feature = geo_people_feature.drop(labels='요일',axis=1)

# 데이터 확인
geo_people_feature
```

```python
geo_people_feature
```

```python
## 주중/주말별 시간대 1~4로 치환
# 주중, 주말로 데이터프레임 나누기
geo_weekday = geo_people_feature[geo_people_feature["주중/주말"] == 0]
geo_weekend = geo_people_feature[geo_people_feature["주중/주말"] == 1]

# 주중 시간대 치환 / 1 : (6, 7, 8, 9, 10), 2: (11, 12, 13, 14), 3: (15, 16, 17, 18), 4: (19, 20, 21, 22, 23)
geo_weekday = geo_weekday.replace([6, 7, 8, 9, 10], 1)
geo_weekday = geo_weekday.replace([11, 12, 13, 14], 2)
geo_weekday = geo_weekday.replace([15, 16, 17, 18], 3)
geo_weekday = geo_weekday.replace([19, 20, 21, 22, 23], 4)

# 주말 시간대 치환 / 1: (7, 8, 9, 10, 11), 2: (12, 13, 14, 15), 3: (16, 17, 18, 19), 4: (20, 21, 22, 23)
geo_weekend = geo_weekend.replace([6, 7, 8, 9, 10, 11], 1)
geo_weekend = geo_weekend.replace([12, 13, 14, 15], 2)
geo_weekend = geo_weekend.replace([16, 17, 18, 19], 3)
geo_weekend = geo_weekend.replace([20, 21, 22, 23], 4)

# 하나의 데이터프레임으로 합치기 및 정렬
geo_total = pd.concat([geo_weekday,geo_weekend], ignore_index=True).sort_values(["매장코드", "기준일ID", "시간대구분"])
geo_total
```

```python
## 최종 feature 생성 
# 각 요일-시간대 별 평균 생활인구수 계산
geo_total_fin = pd.DataFrame({"평균생활인구수" : round(geo_total.groupby(["매장코드", "매장명", "상업지역 면적", "기준일ID", "주중/주말", "시간대구분"])["총생활인구수"].mean(), 2)}).reset_index()

# key(매장코드_영업일자_시간대) 생성
geo_total_fin["매장코드_영업일자_시간대"] = geo_total_fin.apply(lambda x: "%s_%s_%s" %(x["매장코드"], x["기준일ID"].strftime("%Y%m%d"), int(x["시간대구분"])), axis='columns')

geo_total_fin
```

```python
geo_total_fin = geo_total_fin[["매장코드_영업일자_시간대", "매장코드", "매장명", "기준일ID", "주중/주말", "시간대구분", "상업지역 면적", "평균생활인구수"]]
geo_total_fin.to_csv("../../../resources/processedData/매머드_매장별 시간대별 생활인구수.csv", index=False , encoding="utf8")
geo_total_fin
```

```python
## 시간대별 평균 활동인구 당 매출 구하기
sale = pd.read_csv("../../../resources/processedData/매머드_매장별 시간대별 평균매출.csv", index_col =False)
sale
```

```python
## 데이터 합병  
mammoth_feature = pd.merge(geo_total_fin, sale[["매장코드_영업일자_시간대","평균매출"]], on="매장코드_영업일자_시간대", how="inner")

# 생활인구 당 평균 매출 컬럼 생성
mammoth_feature["생활인구당 평균매출"] = round((mammoth_feature["평균매출"]/mammoth_feature["평균생활인구수"])*10000 , 2)
mammoth_feature
```

```python
## 데이터 파싱
mammoth_feature.to_csv("../../../resources/processedData/매머드_feature.csv", encoding="utf8", index=False)
```
