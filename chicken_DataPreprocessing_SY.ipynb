1. 깐부치킨 매장별로 메뉴별로 판매된 개수 카운트 (깐부치킨_매장별_상품별 팔린 개수.csv)

```python
import pandas as pd

data = pd.read_csv('../processedData/깐부치킨_거래데이터.csv', encoding='UTF8')
#이전에 추출했던 데이터를 끌어와서 다시 수정하였음.
#menuData = pd.read_csv('../processedData/깐부치킨_매장별_상품별 팔린 개수.csv', encoding='UTF8')
```

```python
len(data)
```

```python
#매장코드별로 데이터를 가져오고, 상품명별 판매수량 + 소분류명, 상품코드, 판매단가 추가하기
#column -> 매장코드 매장명 소분류명 상품명 상품코드 판매단가 판매개수

CD = sorted(list(set(data['매장코드'])))

menukey = list()    #상품명
menuval = list()    #판매개수

sub = list()    #소분류명
pc = list()     #상품코드
price = list()  #판매단가


for cd in CD :  #매장코드
    df = pd.DataFrame(data[data['매장코드']==cd])  
    
    menu = list(set(df['상품명']))  #해당 매장에서 판매한 상품명 추출
    menulist = dict()               #{ 상품명 : 판매수량 ...} 과 같이 임시로 데이터를 저장할 변수
    
    #상품별로 데이터 저장
    for m in menu :
        menulist[m] = sum(df[df['상품명']==m]['판매수량'])

        sub.append(list(df[df['상품명']==m]['소분류명'])[0])
        pc.append(list(df[df['상품명']==m]['상품코드'])[0])
        price.append(list(df[df['상품명']==m]['판매단가'])[0])

    menukey += list(menulist.keys())
    menuval += list(menulist.values())

#기존의 데이터프레임에 데이터 추가
menuData['소분류명'] = sub
menuData['상품명'] = menukey
menuData['상품코드'] = pc
menuData['판매단가'] = price
menuData['판매개수'] = menuval

```

```python
menuData
```

```python
#정렬

menuData = menuData.sort_values(by=["매장코드", "판매개수", ], ascending=[True, False], ignore_index=True) 
menuData
```

```python
menuData.to_csv('../processedData/깐부치킨_매장별_상품별 팔린 개수.csv', index=False) #저장
```

2. 매장 데이터에서 깐부치킨 매장만 추출 + 영수증 데이터 개수 row 추가 (깐부치킨_매장데이터.csv)

```python
#깐부치킨_매장데이터.csv 추출하기 및 매장별 영수증 데이터 개수 row까지 추가

import pandas as pd


receiptData = pd.read_csv('../processedData/깐부치킨_거래데이터.csv', encoding='UTF8')
storedata_full = pd.read_csv('../rawdata/매장데이터_수정.csv', encoding='UTF8')
storeData = storedata_full[storedata_full['본사코드'] == 'HBAS']    #본사코드가 깐부치킨인 것만 가져오기

sdata = list()   #매장별 영수증 개수를 저장할 리스트
CD = sorted(list(set(storeData['매장코드'])))   #매장코드만 가져오기 
for cd in CD :  #매장코드별(매장별) 행의 개수(영수증의 개수) 추출
    sdata.append(len(receiptData[receiptData['매장코드'] == cd]))

# print(sum(data))   #거래데이터의 row는 83917개. 개수 확인을 위한 코드

storeData['영수증건수'] = sdata
storeData

storeData.to_csv('../processedData/깐부치킨_매장데이터.csv', index=False)
```
