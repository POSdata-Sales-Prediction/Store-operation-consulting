### 매머드커피 주중/주말 매핑 코드 
* 주중 0, 주말 1

```python
## Import modules & Load Dataset
import pandas as pd
mamm = pd.read_csv("../../../resources/extractedData/매머드_거래데이터.csv", encoding="utf8")
```

```python
mamm
```

```python
# 각 지점이 영업일자가 다 동일한지 확인
mamm_code = list(mamm["매장코드"].unique())
for code in mamm_code:
    print("매장명: ",mamm[mamm["매장코드"]==code]["매장명"].values[0],
          ", 영업일자: ",sorted(mamm[mamm["매장코드"]==code]["영업일자"].unique()),", 갯수: ",mamm[mamm["매장코드"]==code]["영업일자"].nunique())
```

```python
mamm["영업일자"] 
```

```python
## 주중, 주말 데이터 매핑하기
# 주중 - 0, 주말 (+[설날] 19-02-04, 19-02-05, 19-02-06) - 1

# 요일 컬럼 추가
import datetime
mamm["영업일자"] = pd.to_datetime(mamm["영업일자"], format="%Y%m%d")

# 0-월, 1-화 ... 5-토,6-일
mamm["요일"] = mamm["영업일자"].dt.weekday

# 요일 > 4 (주말) - 1, 요일 < 5 (주중) - 0
mamm["주중/주말"] = 0
mamm.loc[mamm[mamm["요일"] > 4].index,["주중/주말"]] = 1

# 설날 19-02-04, 19-02-05, 19-02-06 주말 취급
query = (mamm["영업일자"] == "2019-02-04") | (mamm["영업일자"] == "2019-02-05") | (mamm["영업일자"] == "2019-02-06")
idx = mamm[query].index
mamm.loc[idx,["주중/주말"]] = 1

# 요일 컬럼 삭제
mamm = mamm.drop(labels='요일',axis=1)

# 데이터 확인
mamm
```

```python
from collections import Counter

# 주중/주말 별 등장 횟수 카운트
weekend_counter = Counter(mamm['주중/주말'])
weekend_counter
```
