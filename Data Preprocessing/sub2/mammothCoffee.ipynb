```python
## import modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

mamm = pd.read_csv("../../../resources/extractedData/매머드_거래데이터_1차.csv", encoding="utf8")
```

```python
mamm
```

```python
mamm_sorted = mamm[(mamm["소분류명"]!="Add") & (mamm["소분류명"]!="Dessert")]
```

```python
mamm_sorted["결제일시"] = mamm_sorted["결제일시"].astype("str")
mamm_sorted["결제일시"] = pd.to_datetime(mamm_sorted["결제일시"])
```

```python
mamm_sorted["결제일시"]
```

```python
## 판매 매장 갯수 
# mamm_sorted-> 서연이 df로 변경
# 서연이 파일 경로 수정 -> 파일명: 매머드커피_2차_판매 매장 갯수
store_count = pd.DataFrame({'판매매장' : mamm_sorted.groupby(["소분류명","상품명"])["매장코드"].nunique()}).reset_index()
store_count.to_csv("../../../resources/processedData/매머드커피_판매 매장 갯수.csv",index=False, encoding="utf8")
store_count
```

```python
## 1. 매장별 메뉴 판매 갯수 
# mamm_sorted-> 서연이 df로 변경
# 서연이 파일 경로 수정 -> 파일명: 매머드커피_2차_메뉴_판매 개수
menu_store_count = pd.DataFrame({'판매개수' : mamm_sorted.groupby(["상품명","매장코드","매장명"] )["판매수량"].sum()}).reset_index()
menu_store_count.to_csv("../../../resources/processedData/매머드커피_매장별_메뉴_판매 개수.csv",index=False, encoding="utf8")
menu_store_count
```

```python
## 2. 매장별 메뉴 판매 갯수 
# mamm_sorted-> 서연이 df로 변경
# 서연이 파일 경로 수정 -> 파일명: 매머드커피_2차_판매 매장 갯수
menu_store_count = pd.DataFrame({'판매개수' : mamm_sorted.groupby(["상품명","판매단가","매장코드","매장명"] )["판매수량"].sum()}).reset_index()
menu_store_count.to_csv("../../../resources/processedData/매머드커피_매장별_메뉴_판매 개수.csv",index=False, encoding="utf8")
menu_store_count
```

```python
## 3. 메뉴별 총 판매 갯수
# menu_store_count-> 이거 그냥 두면 됨 위에서 만든 데이터 프레임에서 재 추출하는 코드라 가만히 두기!
menu_whole_count = pd.DataFrame({'총판매개수' : menu_store_count.groupby(["상품명"] )["판매개수"].sum()}).reset_index()
menu_whole_count.to_csv("../../../resources/processedData/매머드커피_메뉴별_총 판매 개수.csv",index=False, encoding="utf8")
menu_whole_count
```
