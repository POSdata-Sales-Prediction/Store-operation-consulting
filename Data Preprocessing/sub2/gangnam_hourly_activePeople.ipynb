### 강남구, 송파구, 서초구 시간대별 활동 인구 추출 코드
### feature중 시간대별 활동인구를 추출하기 위함

```python
## Import modules & Load Dataset
import pandas as pd

# 행정동 매핑 데이터 load
code = pd.read_excel("../../../resources/rawdata/행정동코드_매핑정보_20200325.xlsx")

# 생활인구 데이터 load
active = pd.read_csv("../../../resources/rawdata/서울시 활동인구_201902.csv", index_col =False)
```

```python
# 행정동 매핑 데이터 확인 후 column 영문으로 수정 
code 
code.columns = code.iloc[0]
code = code[1:]
code
```

```python
# 강남구, 서초구, 송파구 코드만 sorting
keyword_new = ["강남구","서초구","송파구"]
keyword_new_str = "|".join(keyword_new)

dong_code = code[code["CT_NM"].str.contains(keyword_new_str)]
```

```python
# 생활인구 데이터 확인
active
```

```python
# 생활인구 데이터 중 강남구 데이터만 소팅
active_dong = pd.DataFrame(columns=active.columns)
code = list(dong_code["H_DNG_CD"].astype(int))
for c in code :
    active_dong = active_dong.append(active[active["행정동코드"] == c])
active_dong
```

```python
# 생활 인구 데이터 중 총 생활 인구수만 사용
active_dong = active_dong.iloc[:,[0,1,2,3]]
active_dong
```
