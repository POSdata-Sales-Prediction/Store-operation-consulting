### 매출 데이터 concat 후 특정 점포 매출 동향 파악

Sampled for internal review_transaction_gn_0n_20190805.xlsx 데이터 모두 concat 후 column 명 수정 후 csv로 재추출


```python
# Import Moudules
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# load rawdata
transaction1 = pd.read_excel('../../resources/rawdata/거래데이터01.xls', sheet_name= None)
transaction2 = pd.read_excel('../../resources/rawdata/거래데이터02.xls', sheet_name= None)
transaction3 = pd.read_excel('../../resources/rawdata/거래데이터03.xls', sheet_name= None)
transaction4 = pd.read_excel('../../resources/rawdata/거래데이터04.xls', sheet_name= None)
transaction5 = pd.read_excel('../../resources/rawdata/거래데이터05.xls', sheet_name= None)
transaction6 = pd.read_excel('../../resources/rawdata/거래데이터06.xls', sheet_name= None)
```

```python
transaction1.keys()
```

```python
concatted_tran1 = pd.concat(transaction1, ignore_index=True)
concatted_tran1
```

```python
concatted_tran2 = pd.concat(transaction2, ignore_index=True)
concatted_tran3 = pd.concat(transaction3, ignore_index=True)
concatted_tran4 = pd.concat(transaction4, ignore_index=True)
concatted_tran5 = pd.concat(transaction5, ignore_index=True)
concatted_tran6 = pd.concat(transaction6, ignore_index=True)
```

```python
tran_full = pd.concat([concatted_tran1, concatted_tran2, concatted_tran3, concatted_tran4, concatted_tran5, concatted_tran6])
tran_full
```

```python
# extract full dataset to csv file
tran_full.to_csv('../../resources/rawdata/거래데이터_full.csv', index=False)
```
